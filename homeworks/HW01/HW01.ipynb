```python
import numpy as np
```

```python
# Создание массивов
arr1 = np.array([1, 2, 3, 4, 5])
print ("arr1:", arr1)

arr2 = np.zeros(5)
print("arr2 (zeros):", arr2)

arr3 = np.ones(5)
print("arr3 (ones):", arr3)

arr4 = np.arange(1, 6)
print("arr4 (arange):", arr4)
```

```python
# Проверка формы и типа данных массива
print("Shape of arr1:", arr1.shape)
print("Data type of arr1:", arr1.dtype)
```

```python
# Индексирование и срезы массивов
print("3rd element of arr1:", arr1[2])
print("Slice of arr1 (elements from index 1 to 3):", arr1[1:4])
```

```python
# Векторные операции и базовые статистики
arr5 = arr1 + arr2
print("arr1 + arr2:", arr5)

mean_arr1 = np.mean(arr1)
print("Mean of arr1:", mean_arr1)
```
